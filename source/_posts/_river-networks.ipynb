```python
import numpy as np
from numpy.random import randint, choice
from matplotlib.pyplot import imshow
from matplotlib import cm

class Landscape(object):
    
    def __init__(self, slope, w, h, erosion_depth, avalanche_threshold):
        """
        Parameters
        ----------
        slope: slope of initial featureless incline
        w: length perpendicular to slope
        h: length along slope
        erosion_depth: height eroded by a raindrop
        avalanche_threshold: maximum slope
        """
        
        self._height = slope*np.mgrid[0:h, 0:w][0].astype(float)
        self.erosion_depth = erosion_depth
        self.avalanche_threshold = avalanche_threshold
        self.shape = h, w
        
    def height(self, r):
        "Get the height at one site or at an array of sites."
        try:
            return self._height[zip(*r)]
        except TypeError:
            return self._height[zip(r)]
    
    def erode(self, r):
        "Erode height from one site or from an array of sites."
        try:
            self._height[zip(*r)] -= self.erosion_depth
        except TypeError:
            self._height[zip(r)] -= self.erosion_depth
        
    def avalanche(self):
        "Erode sites with a large gradient."
        stable = [False, False]
        while True:
            for axis in [0, 1]:
                d = np.diff(self._height, axis=axis)
                fall_left = np.roll(d > self.avalanche_threshold, 1)
                fall_right = d < -self.avalanche_threshold
                stable[axis] = ~(fall_left.any() | fall_right.any())
                if stable[axis]:
                    continue  # nothing to do along this axis
                self._height[fall_left] -= -d[fall_left]/4.
                self._height[fall_right] -= -d[fall_right]/4.
            if all(stable):
                break
                
    def plot(self):
        imshow(self._height, cmap=cm.gray)

def raindrop(landscape, T, initial_site):
    "Place a raindrop and return its path over the landscape."
    prev_site = None
    site = np.asarray(initial_site)
    sites = [site]
    while sites[-1][0] != 0:
        next_site = trickle(landscape, prev_site, site, T)
        sites.append(next_site)
        site, prev_site = next_site, site
    return sites

def trickle(landscape, prev_site, site, T):
    "Advance a raindrop one step."
    neighbors = np.mod(site - [[-1, 0], [1, 0], [0, -1], [0, 1]], landscape.shape)
    dh = landscape.height(site) - landscape.height(neighbors)
    not_uphill = dh >= 0
    # Does the raindrop have no choice but to backtrack?
    if prev_site is not None:
        fresh_sites = np.any(neighbors != prev_site, -1)
        if not np.any(not_uphill[fresh_sites]):
            return prev_site
    # Randomly choose among non-backtracking neighbors.
    boltzman_weights = np.exp(dh/T)*not_uphill
    p = boltzman_weights/boltzman_weights.sum()
    return neighbors[choice([0, 1, 2, 3], p=p)]

def rain(landscape, T, N):
    "Randomly place raindrops, eroding and avalanching after each drop."
    for _ in xrange(N):
        sites = raindrop(landscape, T, map(randint, landscape.shape))
        landscape.erode(sites)
        landscape.avalanche()
```

Next, characterize the river network by dropping rain drops at each point and marking sites traversed by more than R=100 raindrops.

```python
def define_network(landscape, T):
    network = np.zeros_like(landscape)
    all_sites = np.indices(landscape.shape).reshape(2, -1)[network.shape[0]:]
    for initial_site in all_sites:
        sites = raindrop(network, T, initial_site)
        network[zip(*sites)] += 1
    return network
```

```python
define_network(a, 20)
```
